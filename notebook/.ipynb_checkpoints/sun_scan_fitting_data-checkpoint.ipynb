```python
import pickle
import matplotlib.pyplot
import numpy
import scipy.optimize
import pathlib
import multiprocessing
import zipfile
```

```python
tp_topics = [
        'xffts_power_board01',
        'xffts_power_board02',
        'xffts_power_board03',
        'xffts_power_board04',
        'xffts_power_board05', 
        'xffts_power_board06',
        'xffts_power_board07',
        'xffts_power_board08',
        'xffts_power_board09',
        'xffts_power_board10', 
        'xffts_power_board11',
        'xffts_power_board12',
        'xffts_power_board13', 
        'xffts_power_board14', 
        'xffts_power_board15',
        'xffts_power_board16',
]
```

```python
def differentiation(data):
    ta = {}
    for _ in data:
        ta[_] = data[_]['ta']
        
    dta = {
        _ : ta[_].values[1:] - ta[_].values[:-1]
        for _ in tp_topics
        }
    return dta


def gauss(x, *p):
    A, mu, sigma = p
    return A*numpy.exp(-(x-mu)**2/(2.*sigma**2))


def daz_gauss_fitting(data, dta, gauss):
    a_co1= {}
    a_co2 = {}
    daz = {}
    for _ in tp_topics:
        daz[_] = data[_]['daz']*numpy.cos(numpy.deg2rad(data[_]['el']))
    
    for _ in tp_topics:
        try:
            _x1 = daz[_][daz[_]<0][:-1]
            _y1 = dta[_][daz[_][:-1]<0]
            a_co1[_], var1 = scipy.optimize.curve_fit(gauss, _x1, _y1, p0=[1, 0, 1])
            if abs(a_co1[_][0]) > 10000:
                a_co1[_] = [numpy.nan, numpy.nan, numpy.nan]
            if abs(a_co1[_][1]) > 1:
                a_co1[_] = [numpy.nan, numpy.nan, numpy.nan]
            if abs(a_co1[_][2]) > 100:
                a_co1[_] = [numpy.nan, numpy.nan, numpy.nan]        
        
        except (KeyError, TypeError, RuntimeError, ValueError):
            a_co1[_] = [numpy.nan, numpy.nan, numpy.nan]
            pass           
        
        try:
            _x2 = daz[_][daz[_]>0][:-1]
            _y2 = dta[_][daz[_][1:]>0]
            a_co2[_], var2 = scipy.optimize.curve_fit(gauss, _x2, _y2, p0=[1, 0, 1])        
            if abs(a_co2[_][0]) > 10000:
                a_co2[_] = [numpy.nan, numpy.nan, numpy.nan]
            if abs(a_co2[_][1]) > 1:
                a_co2[_] = [numpy.nan, numpy.nan, numpy.nan]
            if abs(a_co2[_][2]) > 100:
                a_co2[_] = [numpy.nan, numpy.nan, numpy.nan]
                
        except (KeyError, TypeError, RuntimeError, ValueError):
            a_co2[_] = [numpy.nan, numpy.nan, numpy.nan]
            pass

    return a_co1,a_co2, 

def del_gauss_fitting(data, dta, gauss):
    e_co1= {}
    e_co2 = {}
    del_ = {}
    for _ in tp_topics:
        del_[_] = data[_]['del']
    
    for _ in tp_topics:
        try:
            _x1 = del_[_][del_[_]<0][:-1]
            _y1 = dta[_][del_[_][:-1]<0]
            e_co1[_], var1 = scipy.optimize.curve_fit(gauss, _x1, _y1, p0=[1, 0, 1])
            if abs(e_co1[_][0]) > 10000:
                e_co1[_] = [numpy.nan, numpy.nan, numpy.nan]
            if abs(e_co1[_][1]) > 1:
                e_co1[_] = [numpy.nan, numpy.nan, numpy.nan]
            if abs(e_co1[_][2]) > 100:
                e_co1[_] = [numpy.nan, numpy.nan, numpy.nan]
                
        except (KeyError, TypeError, RuntimeError, ValueError):
            e_co1[_] = [numpy.nan, numpy.nan, numpy.nan]
            pass
        
        try:
            _x2 = del_[_][del_[_]>0][:-1]
            _y2 = dta[_][del_[_][1:]>0]
            e_co2[_], var2 = scipy.optimize.curve_fit(gauss, _x2, _y2, p0=[1, 0, 1])
            if abs(e_co2[_][0]) > 10000:
                e_co2[_] = [numpy.nan, numpy.nan, numpy.nan]
            if abs(e_co2[_][1]) > 1:
                e_co2[_] = [numpy.nan, numpy.nan, numpy.nan]
            if abs(e_co2[_][2]) > 100:
                e_co2[_] = [numpy.nan, numpy.nan, numpy.nan]
                
        except (KeyError, TypeError, RuntimeError, ValueError):
            e_co2[_] = [numpy.nan, numpy.nan, numpy.nan]
            pass

    return e_co1,e_co2, 


def get_center(a_co1, a_co2, e_co1, e_co2):
    az_center = {}
    el_center = {}
    
    for _ in a_co2:
        try:
            az_center[_] = (a_co1[_][1]+a_co2[_][1])/2
        except:
            az_center[_] = numpy.nan
            pass
        
        try:
            el_center[_] = (e_co1[_][1]+e_co2[_][1])/2
        except:
            el_center[_] = numpy.nan
            pass
    return az_center, el_center


def get_hpbw(a_co1, a_co2, e_co1, e_co2):
    az_hpbw = {}
    el_hpbw = {}
    
    for _ in a_co2:
        try:
            az_hpbw[_] = numpy.mean([abs(a_co1[_][2]), abs(a_co2[_][2])])*2.3548
        except:
            az_hpbw[_] = numpy.nan
            pass
        
        try:
            el_hpbw[_] = numpy.mean([abs(e_co1[_][2]), abs(e_co2[_][2])])*2.3548
        except:
            el_hpbw[_] = numpy.nan
            pass
    return az_hpbw, el_hpbw


def az_el_average(data):
    az = numpy.mean(data['xffts_power_board01']['az'])
    el = numpy.mean(data['xffts_power_board01']['el'])
    return az, el    


def get_timestamp(data):
    time = str(numpy.median(data['xffts_power_board01'].index.values.astype('int64')).astype('datetime64[ns]'))
    return time

def get_sun_size(a_co1, a_co2, e_co1, e_co2):
    az_sun_size = {}
    el_sun_size = {}
    sun_size = {} 
    for _ in a_co2:
        try:
            az_sun_size[_] = abs(a_co1[_][1]) + abs(a_co2[_][1])
        except:
            az_sun_size[_] = numpy.nan
            pass
        
        try:
            el_sun_size[_] = abs(e_co1[_][1]) + abs(e_co2[_][1])
        except:
            el_sun_size[_] = numpy.nan
            pass
        
        sun_size[_] = numpy.nanmean([az_sun_size[_], el_sun_size[_]])
        
    return az_sun_size, el_sun_size, sun_size
    

def judge_direction(az_center, el_center):
    direction = ''
    if az_center['xffts_power_board01'] == az_center['xffts_power_board01']:
        direction = 'az'
    elif el_center['xffts_power_board01'] == el_center['xffts_power_board01']:
        direction = 'el'
    
    return direction


def writeto(tables, path):
    f = open(path, 'wb')
    pickle.dump(tables, f)
    f.close()
    return
```

```python
saveto = '/home/amigos/data/otf_planet2018/moon_scan/20200317/fitting_data/'

def get_fitting_data(path):
    if path.name.find('moon') == -1: return
    try:
        f = open(path, 'rb')
        data = pickle.load(f)
        print(path)
        dta = differentiation(data)
        a_co1, a_co2 = daz_gauss_fitting(data, dta, gauss)
        e_co1, e_co2 = del_gauss_fitting(data, dta, gauss)
        az_center, el_center = get_center(a_co1,a_co2, e_co1, e_co2)
        az_hpbw, el_hpbw = get_hpbw(a_co1,a_co2, e_co1, e_co2)
        az_sun_size, el_sun_size, sun_size = get_sun_size(a_co1,a_co2, e_co1, e_co2)
        az, el = az_el_average(data)
        timestamp = get_timestamp(data)
        direction =  judge_direction(az_center, el_center)
        print(direction)
        tables = {
            'az_hpbw': az_hpbw,
            'el_hpbw': el_hpbw,
            'az_center': az_center,
            'el_center': el_center,
            'az_average': az,
            'el_average': el,
            'az_coeff': (a_co1, a_co2),
            'el_coeff': (e_co1, e_co2),
            'timestamp': timestamp,
            'az_sun_size': az_sun_size,
            'el_sun_size': el_sun_size,
            'sun_size': sun_size,
            'direction': direction
        }
        writeto(tables, saveto + path.name.strip('.pickle') + '_fitting.pickle' )
        f.close()
        print('ok')
    except ValueError:
        pass
    return 
#いるのは、az,el,daz_center,del_center,hpbw
```

```python
data_file = pathlib.Path('/home/amigos/data/otf_planet2018/moon_scan/20200317/chopper_data/').iterdir()

with multiprocessing.Pool(processes=3) as pool:
    pool.map(get_fitting_data, data_file)
    pass
```
